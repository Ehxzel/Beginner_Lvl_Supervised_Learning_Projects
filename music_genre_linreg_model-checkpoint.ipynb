```python
#CReating MultiClass Classification Model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split, KFold, cross_val_score
import pandas as pd
import numpy as np

music_df = pd.read_csv("MusicCharDataset.csv")

#Handling Missing Values
music_df = music_df.fillna(music_df.mode().iloc[0])

# Label encode the 'Genre' column
le = LabelEncoder()
music_df["genre"] = le.fit_transform(music_df["genre"].astype(str))
music_df["genre"] = music_df["genre"].apply(pd.to_numeric)
# Label encode the 'Filename' column
le = LabelEncoder()
music_df["filename"] = le.fit_transform(music_df["filename"].astype(str))
music_df["filename"] = music_df["filename"].apply(pd.to_numeric)
# Taagets amd Features
X = music_df.drop("genre", axis=1 ).values
y = music_df["genre"].values

#train our model
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#Setting up Dictionaries
train_accuracies = {}
test_accuracies = {}

#loading Linear Reg anmd finding the best alpha for our model
models = {"ridge": Ridge(alpha = 0.16), "knn": KNeighborsClassifier(n_neighbors = 15), "tree": DecisionTreeClassifier()}
results = []
for mod in models.values():
    #Perfoming KF and Cross Validation
    kf = KFold(n_splits =3 , shuffle = True, random_state = 42)
    cv_scores = cross_val_score(mod,X_train, y_train, cv=kf, scoring ="neg_mean_squared_error")
    results.append(cv_scores)
#Computing the Metrics
rmse = np.sqrt(-cv_scores)    
Average_RMSE = np.mean(rmse)
stan_dev = np.std(y)
    
print("Cross Validation Scores for the models: ", results, "\n")
print("RMSE Value for the models: ", rmse, "\n")
print("Average RMSE Value for the models: ", Average_RMSE, "\n")
print("Standard Deviation for each Score:", stan_dev, "\n")
```
